```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
```

```python
# Create city weather dataframe from csv file
city_weather_df = pd.read_csv('../output_data/ML_city_weather.csv')

city_weather_df.head()
```

```python
# Create a Scatter Plot for checking the coverage of the world cities
x_values = city_weather_df['Longitude']
y_values = city_weather_df['Latitude']
plt.scatter(x_values,y_values, edgecolors="black")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.title("City Longitude vs Latitude (04/12/20)")
#plt.savefig(f"../output_data/ML_fig1.png")
plt.show()
```
